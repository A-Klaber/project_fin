```python
# Files to Load
alt_stat_0522 = pd.read_csv("./alt_fuel_stations (May 19 2022).csv",low_memory=False)
```

```python
alt_stat_0522_cleaner = alt_stat_0522.drop(['Country','Fuel Type Code','Station Phone','Intersection Directions','Street Address','Federal Agency Code','Groups With Access Code (French)','Federal Agency Code','Access Detail Code',
'Federal Agency Name','Federal Agency ID'],axis = 1)
alt_stat_0522_cleaner
```

```python
#change positive Longitude to negative
alt_stat_0522_cleaner.loc[alt_stat_0522_cleaner['Longitude'] > 20,'Longitude'] *= -1

# delete rows where lat = zero
alt_stat_0522_cleaner = alt_stat_0522_cleaner[alt_stat_0522_cleaner.Longitude != 0]
    
    
```
